```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
#plt.rcParams['font.family'] = 'Times New Roman'
```

```python
material=['KVBaCr','CaVBaCr','KVBiCr','KVSrCr','KVCaCr']
stability_query=[1.271,0.93,1.081,1.23,1.273]
stability_query=np.array(stability_query)
stability_predict=[1.186,0.9341,1.1851,1.1419,1.1592]
stability_predict=np.array(stability_predict)
error_precent=(stability_predict-stability_query)/stability_query*100
```

```python
bar_width = 0.2
space=0.15
index = np.arange(len(material))
fig, ax = plt.subplots(figsize=(5,3.5))
bars1 = ax.bar(material, stability_query, bar_width, label='database')
bars2 = ax.bar(index + bar_width+space, stability_predict, bar_width, label='Attention CNN')
ax.set_ylabel(r'$\mathrm{E_{hull}\ (eV/atom)}$')

#for i in range(len(material)):
#    ax.text(i+(bar_width+space)/2, max(stability_predict[i],stability_query[i])+0.03, '%1.2f%%'%(error_precent[i]), ha='center', va='bottom')

ax.set_xticks(index + (bar_width+space) / 2)
ax.set_xticklabels(material)
ax.legend(frameon=False)
#ax.set_aspect('equal')
ax.yaxis.set_major_locator(MaxNLocator(nbins=6))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis='y', direction='in')
ax.tick_params(axis='y', which='minor', direction='in')
ax.set_ylim([0,1.65])
plt.savefig('../result/stability_contrast.png',dpi=2400)
```

```python
material=['FeCoIr','FeNiGeRh','ScNiCrRh','FeCoZrIr','FeRhZn']
mag_predict=[1.221406,1.118848,1.248265,1.095930,1.095549]
mag_predict=np.array(mag_predict)
mag_cal=[4.8664,4.8609,4.1014,3.9048,3.8993]
mag_cal=np.array(mag_cal)
E0_predict=[-7.905418,-6.652576,-7.527874,-8.513762,-5.761688]
E0_predict=np.array(E0_predict)
E0_cal=[-22.776/3,-26.188/4,-29.594/4,-33.683/4,-15.900/3]
E0_cal=np.array(E0_cal)
```

```python
bar_width = 0.2
space=0.15
index = np.arange(len(material))
fig, ax = plt.subplots(figsize=(5,3.5))
bars1 = ax.bar(material, mag_predict, bar_width, label='Predict')
bars2 = ax.bar(index + bar_width+space, mag_cal/4, bar_width, label='Calculated')
ax.set_ylabel(r'$\mathrm{m_s\ (\mu_B/atom)}$')

ax.set_xticks(index + (bar_width+space) / 2)
ax.set_xticklabels(material)
ax.legend(frameon=False)
#ax.set_aspect('equal')
ax.yaxis.set_major_locator(MaxNLocator(nbins=6))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis='y', direction='in')
ax.tick_params(axis='y', which='minor', direction='in')
ax.set_ylim([0,1.5])
plt.savefig('../result/mag_contrast.png',dpi=2400)
```

```python
bar_width = 0.2
space=0.15
index = np.arange(len(material))
fig, ax = plt.subplots(figsize=(5,3.5))
bars1 = ax.bar(material, -E0_predict, bar_width, label='Attention CNN')
bars2 = ax.bar(index + bar_width+space, -E0_cal, bar_width, label='first-principles calculations')
ax.set_ylabel(r'$\mathrm{-E_0\ (eV/atom)}$')

ax.set_xticks(index + (bar_width+space) / 2)
ax.set_xticklabels(material)
ax.legend(frameon=False,loc='upper left')
#ax.set_aspect('equal')
ax.yaxis.set_major_locator(MaxNLocator(nbins=6))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis='y', direction='in')
ax.tick_params(axis='y', which='minor', direction='in')
ax.set_ylim([0,12])
plt.savefig('../result/E0_contrast.pdf')
```
